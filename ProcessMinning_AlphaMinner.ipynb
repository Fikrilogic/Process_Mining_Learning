```python
import pm4py as pm
from pm4py.statistics.traces.generic.pandas import case_statistics
```

```python
data = data.rename(columns={'case_id': 'case:concept:name', 'activity':'concept:name', 'timestamp': 'time:timestamp'})
start = pm.get_start_activities(data)
df_start_activities = pm.filter_start_activities(data,['register request'])
```

```python
df_start_activities
```

```python
end_activities = pm.get_end_activities(df_start_activities)
df_filtered_end = pm.filter_end_activities(df_start_activities, end_activities)
```

```python
end_activities
```

```python
df_filtered_end
```

```python
from pm4py.objects.conversion.log import converter as log_converter
log = log_converter.apply(df_filtered_end)
log
```

```python
from pm4py.algo.discovery.alpha import algorithm as alpha_miner
net, initial_marking, final_marking = alpha_miner.apply(log)
```

```python
from pm4py.visualization.petri_net import visualizer as pn_visualization
gviz = pn_visualization.apply(net, initial_marking, final_marking)
pn_visualization.save(gviz, 'pn.png')
```
